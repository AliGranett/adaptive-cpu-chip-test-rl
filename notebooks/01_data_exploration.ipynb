# 01 - Data Exploration

Overview of the Stage-2 CPU chip-testing dataset: schema, label balance, missing values and feature/label relationships. If no raw CSV exists, a synthetic dataset is generated automatically.

```python
import sys, pathlib
ROOT = pathlib.Path.cwd()
while not (ROOT / 'src').exists() and ROOT != ROOT.parent:
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT))

import pandas as pd
import matplotlib.pyplot as plt
from src.config import CONFIG
from src.data.loader import load_raw_dataset
from src.data.feature_engineering import engineer_features

df = load_raw_dataset(CONFIG)
df.head()
```

```python
print('Shape:', df.shape)
print('\nLabel distribution:')
print(df[CONFIG.env.label_column].value_counts(normalize=True))
print('\nMissing values per column (top 10):')
print(df.isna().sum().sort_values(ascending=False).head(10))
```

```python
df.describe().T.head(15)
```

```python
# Correlation of measurements with the failure label.
eng = engineer_features(df, CONFIG)
label = CONFIG.env.label_column
corr = eng.corr(numeric_only=True)[label].drop(label).sort_values()
ax = corr.plot(kind='barh', figsize=(7, 8), title='Feature correlation with FAIL label')
ax.set_xlabel('Pearson correlation')
plt.tight_layout()
plt.show()
```
